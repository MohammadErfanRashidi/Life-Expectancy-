```python
# Standardize columns names 
df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]
print(df.columns.tolist())
```

```python
# Converting 'status' col into numerical
print(df['status'].unique())

df['status'] = df['status'].map({'Developing': 0, 'Developed': 1})

print(df['status'].unique())
```

```python
# Check missing values per rows with more than 90 percent missing
(df.drop(['country', 'year', 'status'], axis = 1).isnull().mean(axis = 1) > 0.9).sum()
```

```python
# Sort the data for time-based operations
df = df.sort_values(['country', 'year']).reset_index(drop = True)
```

```python
# Check if missing values have a meaning or just random
df.groupby(df['gdp'].isnull().astype(int))['life_expectancy'].mean()
```

```python
# Drop the missing values in Life expectancy due to it being our target
df = df.dropna(subset = ['life_expectancy'])
```

```python
# Fill missing values based on the skew 
feature_cols = [
    'adult_mortality', 'infant_deaths', 'alcohol', 'percentage_expenditure',
    'hepatitis_b', 'measles', 'bmi', 'under-five_deaths', 'polio',
    'total_expenditure', 'diphtheria', 'hiv/aids', 'gdp', 'population',
    'thinness__1-19_years', 'thinness_5-9_years',
    'income_composition_of_resources', 'schooling'
]

method = {}
for col in feature_cols:
    col_data = df[col].dropna()
    if len(col_data) > 0:
        sk = skew(col_data)
        # If it is less, we have to mean
        method[col] = abs(sk) >= 1
    else:
        # Default it is going to median
        method[col] = True 

# Display the methods
for key, value in method.items():
    if ~value:
        print(f'{key}: {value}')
```

We are going to use mean for the values above.

```python
# Fill based on country, status, and global
country_stats = {}
status_stats = {}
global_stats = {}

for col in feature_cols:
    agg_func = 'median' if method[col] else 'mean'

    global_stats[col] = df[col].agg(agg_func)

    status_stats[col] = df.groupby('status')[col].agg(agg_func)

    country_stats[col] = df.groupby('country')[col].agg(agg_func)

# Apply imputation 
for col in feature_cols:
    # 1. Fill with country 
    df[col] = df[col].fillna(
        df['country'].map(country_stats[col])
    )
    # 2. Fill remaining with status
    df[col] = df[col].fillna(
        df['status'].map(status_stats[col])
    )
    # 3. Fill any still remaining with global stat
    df[col] = df[col].fillna(global_stats[col])

df[feature_cols].isnull().sum().sum()
```

```python
# IQR func to detect outliers
def detect_outliers_iqr(df, columns = None):
    '''
    Returns a boolean DataFrame of the same shape, where True marks outliers 
    (values outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR]).
    
    Parameters
    ----------
    df : pandas DataFrame
    columns : list of str, optional
        Columns to check. If None, all numeric columns are used.
    
    Returns
    -------
    pandas DataFrame (boolean)
    '''
    if columns is None:
        columns = df.select_dtypes(include = 'number').columns

    outlier_mask = pd.DataFrame(False, index = df.index, columns = df.columns)

    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - (1.5 * IQR)
        upper = Q3 + (1.5 * IQR)
        outlier_mask[col] = (df[col] < lower) | (df[col] > upper)

    return outlier_mask
```

```python
# Detect outliers
outlier_flag = detect_outliers_iqr(df, columns = feature_cols)

# Count the outliers
outlier_counts = outlier_flag.sum().sort_values(ascending = False)
print(outlier_counts[outlier_counts > 0])
```
